** 2. Keyword arguments**

In short, a keyword argument is an argument
which is supplied to the function by explicitly naming each parameter
and specifying its value. Often used with plt are **linewidth** and **markersize**.

```python
# Example of plot using different markers specified in plt.plot() method

x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

plt.plot(x,y1, "bo-")
plt.show()
```

```python
# Example of plot using different linewidth, markersize, and markers
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

plt.plot(x,y1, "bo-", linewidth = 2, markersize= 12)
plt.show()
```

```python
# Example 2 of plot using different linewidth, markersize, and markers
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

plt.plot(x,y2, "gs-", linewidth = 2, markersize= 12)
plt.show()
```

## Video 2.3.2: Customizing Your Plots

```python
#Import:
import numpy as np
import matplotlib.pyplot as plt
```

There are a few important elements that can be easily added to plots. Here are some of them

**legent()** : add legend  
**axis()** : ajust axis  
**xlabel(); ylabel()** : set x and y axis lables  
**savefig()** : save figure as (extension of the file shoud be specifie as a method argument).

```python
# Generate data to plot:
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

# plot the data with addition of xlables, ylables
plt.plot(x,y1, "bo:", linewidth = 2, markersize= 4)
plt.plot(x,y2, "gs--", linewidth = 2, markersize= 6)
plt.xlabel("X"); plt.ylabel("Y")
plt.show()
```

**PLT** suports LaTeX syntax !

```python
# Generate data to plot:
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

# plot the data with addition of xlables, ylables and specified axis
plt.plot(x,y1, "bo:", linewidth = 2, markersize= 4)
plt.plot(x,y2, "gs--", linewidth = 2, markersize= 6)
plt.xlabel("$X$")
plt.ylabel("$Y$") # $ LaTeX syntax 
plt.axis([-0.5, 10.5, -10, 110]) # in the format of plt.axis([xmin, xmax, ymin, ymax])

plt.show()
```

To use lable() methos we have to add lable keyword lable to the plot definition 

```python
# All together:

# Import:
import numpy as np
import matplotlib.pyplot as plt

# Generate data:
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

#Plot, lables, axis, legend, and savefig:
plt.plot(x,y1, "bo:", linewidth = 2, markersize= 4, label = "First")
plt.plot(x,y2, "gs--", linewidth = 2, markersize= 6, label = "Second")
plt.xlabel("$X$")
plt.ylabel("$Y$") # $ LaTeX syntax 
plt.axis([-0.5, 10.5, -10, 110]) # in the format of plt.axis([xmin, xmax, ymin, ymax])
plt.legend(loc = "upper left") # add legend (lable in plt.plot is necesery)
plt.savefig("plotTest.png") # save plot as .png

plt.show()
```

## Video 2.3.3: Plotting Using Logarithmic Axes

The **plt** functions that are used to make logarithmic plots are as follows:  
**semilogx()** plots the x-axes on a log scale and the y in the original scale.  
**semilogy()** plots the y-axes on the log scale and the x in the original scale.  
**loglog()** plots both x and y on logarithmic scales.  

```python
# Import:
import numpy as np
import matplotlib.pyplot as plt

# Generate data:
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5

#Plot usinf loglog() method - logarythmic plot
plt.loglog(x,y1, "ro:", linewidth = 2, markersize= 4, label = "First log")
plt.loglog(x,y2, "ys--", linewidth = 2, markersize= 6, label = "Second log")
plt.xlabel("log X")
plt.ylabel("log Y") # $ LaTeX syntax 
# plt.axis([-0.5, 10.5, -10, 110]) # in the format of plt.axis([xmin, xmax, ymin, ymax])
plt.legend(loc = "upper left") # add legend (lable in plt.plot is necesery)
# plt.savefig("plotTest.png") # save plot as .png

plt.show()
```

## 2.3.4: Generating Histograms
Learn how to generate histograms using **plt.hist()**  
Learn how to create subplots using **plt.subplot()**

```python
# Import:
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data:
x = np.random.normal(size=1000)

#plot histogram:
plt.hist(x);
```

By default, hist uses 10 evenly spaced bins
and it tries to optimize both bin width and bin locations.

```python
# plot normed histogram:
plt.hist(x, normed = True);
```

Instead of having the number of observations that fall in each bin,
we have the proportion of observations that fall in each bin.  
That's what it means for a histogram to be normalized.

```python
# plot normed histogram with defined bins
plt.hist(x, normed = True, bins = np.linspace(-5,5,21));
```

In this case, you see that the histogram looks different.
That's because we have specified 20 bins between the numbers minus 5 and plus 5.

**Let's make sure we understand why.
If we think about the first example where we just
have one bin, to have one bin we need to specify
the start point-- the start location of the bin
and the end location of that bin.
So to get one bin, we need two points.
If we had two bins we would need three points, and so on.
This is why to have 20 bins, we will need to have 20 plus 1,
or 21, points along the x-axis.**

Let's then examine another distribution, which is a bit more exotic,
the so-called **gamma distribution**.  
It is a continuous probability density function that starts at 0
and goes all the way to positive infinity.The gamma distribution, like the normal distribution, has two parameters.
For now, we don't need to be too concerned about the nature of the gamma
distribution.
We just know that it's some type of probability distribution,
and we'd like to investigate its shape using histograms.

We'll also meet here a plt function called
subplot, **plt.subplot()** which enables us to have several subplots within each figure.  
This upload function takes in three arguments where the first two specify
the number of rows and the number of columns in the subplot,
and the third argument gives the plot number.
For example, if you specify two rows and three columns,
then you will have six subplots.The plot number always starts at 1, so in a two by three subplot,
the plot numbers range from 1 to 6, where the plot numbers are incremented across rows first.

```python
x = np.random.gamma(2,3,100000)
plt.hist(x, bins = 30);
```

```python
plt.hist(x, bins = 30, normed = True);
```

```python
plt.hist(x, bins = 30, cumulative = True);
```

```python
plt.hist(x, bins = 30, normed = True, cumulative = True);
```

**We can now pool all of these four different histograms into one figure.**

```python
#Sub plot examplr - many plots on the same display space:

plt.figure()
plt.subplot(221) # subplot 2 raws, 2 columns , position 1
plt.hist(x, bins = 30);
plt.subplot(222) # subplot 2 raws, 2 columns , position 2
plt.hist(x, bins = 30, normed = True);
plt.subplot(223) # subplot 2 raws, 2 columns , position 3
plt.hist(x, bins = 30, cumulative = True);
plt.subplot(224) # subplot 2 raws, 2 columns , position 4
plt.hist(x, bins = 30, normed = True, cumulative = True, histtype="step");
```
